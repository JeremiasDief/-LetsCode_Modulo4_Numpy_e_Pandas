## Exercício 2

**a)** Utilizando a base de dados de covid, crie um gráfico de linhas mostrando a evolução do número de casos novos mês a mês, para as DRS 'Grande São Paulo', 'São José do Rio Preto' e 'Marília'. Identifique os DRS por cores e legenda. Não esqueça de colocar título para o gráfico e para os eixos. 

```python
df2
```

```python
df2_pivotado = df2.pivot_table(index="ano_mes", columns="nome_drs", values="casos_novos", aggfunc="sum")
df2_pivotado = df2_pivotado[['Grande São Paulo', 'São José do Rio Preto', 'Marília']]
df2_pivotado
```

```python
ax = df2_pivotado.plot(kind="line", title="Evolução dos novos casos de covid em 3 regiões de SP", figsize=(5,5),
                 xlabel="Ano e mês", ylabel="Número de novos casos")
ax.legend(title="Regiões")
```

**b)** Repita o gráfico do item a trocando as cores dos DRS.

```python
ax = df2_pivotado.plot(kind="line", title="Evolução dos novos casos de covid em 3 regiões de SP", figsize=(5,5),
                 xlabel="Ano e mês", ylabel="Número de novos casos", color=["red", "black", "yellow"])
ax.legend(title="Regiões")
```

**c)** Considerando apenas os DRS 'Bauru', 'São José do Rio Preto' e 'Marília', faça três gráficos de pizza, para os meses de abril, maio e junho, mostrando a porcentagem de novos óbitos em cada DRS.

```python
df2_pivotado2 = df2.pivot_table(index="ano_mes", columns="nome_drs", values="obitos_novos", aggfunc="sum")
df2_pivotado2 = df2_pivotado2[2:][['Bauru', 'São José do Rio Preto', 'Marília']]
df2_pivotado2
```
